```python
key='Sex'
col='green'
grouped2=df.groupby([key]).mean()
fig, ax=plt.subplots(nrows=2, ncols=2)
ax[0,0].bar(grouped2.index,grouped2['Salary'],color=col,hatch='/')
ax[0,0].set_ylabel('Average salary')
ax[0,1].bar(grouped2.index,grouped2['EngagementSurvey'],color=col,hatch='o')
ax[0,1].set_ylabel('Average engagement survey')
ax[1,0].bar(grouped2.index,grouped2['EmpSatisfaction'],color=col,hatch='+')
ax[1,0].set_ylabel('Average employee satisfaction')
ax[1,1].bar(grouped2.index,grouped2['Absences'],color=col,hatch='*')
ax[1,1].set_ylabel('Average absences')
fig.suptitle('Under different sex')
```

```python
key='MaritalDesc'
col='purple'
grouped2=df.groupby([key]).mean()
fig, ax=plt.subplots(nrows=2, ncols=2)
ax[0,0].barh(grouped2.index,grouped2['Salary'],color=col,hatch='/')
ax[0,0].set_ylabel('Average salary')
ax[0,1].barh(grouped2.index,grouped2['EngagementSurvey'],color=col,hatch='o')
ax[0,1].set_ylabel('Average engagement survey')
ax[0,1].yaxis.set_label_position("right")
ax[0,1].yaxis.tick_right()
ax[1,0].barh(grouped2.index,grouped2['EmpSatisfaction'],color=col,hatch='+')
ax[1,0].set_ylabel('Average employee satisfaction')
ax[1,1].barh(grouped2.index,grouped2['Absences'],color=col,hatch='*')
ax[1,1].set_ylabel('Average absences')
ax[1,1].yaxis.set_label_position("right")
ax[1,1].yaxis.tick_right()
fig.suptitle('Under different marital status')
```
